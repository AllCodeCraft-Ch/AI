```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from matplotlib import pyplot as plt
 
train = pd.read_csv('StudentScore.Train.csv') # , header=None
test = pd.read_csv('StudentScore.Test.csv')
 
X_train = np.asarray(train.iloc[:,:-2]).sum(axis=1).reshape(-1,1)
y_train = np.asarray(train.iloc[:,-1]).reshape(-1,)
X_test = np.asarray(test.iloc[:,:-2]).sum(axis=1).reshape(-1,1)
y_test = np.asarray(test.iloc[:,-1]).reshape(-1,)
 
np.random.seed(10)
model = MLPClassifier(activation='identity', solver='lbfgs',
    hidden_layer_sizes=(1,10,), max_iter=1000, early_stopping=True)
model = MLPClassifier(activation='relu', solver='sgd', learning_rate_init=1e-3,
                     hidden_layer_sizes=(1,10,), max_iter=10000, early_stopping=True)
 
model.fit(X_train, y_train)
 
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}%')
print(y_pred)

plt.plot(model.loss_curve_)
```
